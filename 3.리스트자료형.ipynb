```python
#리스트는 어떻게 만들고 사용할까?
#리스트명 = [요소1, 요소2, 요소3, ...]

a=[]
b=[1,2,3]
c=['Life', 'is', 'too', 'short']
d=[1,2,'Life', 'is']
e=[1,2,['Life', 'is']]
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
#리스트의 인덱싱과 슬라이싱
a=[1,2,3]
print(a[0])
print(a[2])
print(type(a[2]))
```

```python
print(a[0]+a[2])
print(a[-3])
```

```python
b=['1', '2', '3','a','b','c']
print(b[0]+b[2]+b[-1])
```

```python
c=[1,2,['a','b',['Life', 'is']]]
print(len(c)) #리스트가 몇개인가
c[-1][-1][0] #리스트 안에 리스트의 요소를 인덱싱하기
```

```python
#리스트의 슬라이싱
a=[1,2,3,4,5]
print(a[0:2]) #0부터 2까지 (2는 포함하지 않음)
print(a[2:5]) #2부터 5까지 (5는 포함하지 않음)
print(a[:2]) #처음부터 2까지
print(a[2:5]) #2부터 5까지 (5는 포함하지 않음)
```

```python
print(a[:]) #처음부터 끝까지
print(a[::2]) #처음부터 끝까지 2칸씩 띄어서
print(a[::3]) #처음부터 끝까지 3칸씩 띄어서
print(a[::]) #처음부터 끝까지 1칸씩 띄어서
print(a[1::2]) #1부터 끝까지 2칸씩 띄어서
print(a[1:4:2]) #1부터 4까지 2칸씩 띄어서
print(a[1:4:3]) #1부터 4까지 3칸씩 띄어서
print(a[1:4:1]) #1부터 4까지 1칸씩 띄어서
```

```python
#리스트 더하기 (+)
a=[1,3,5]
b=[2,4,6]
print(a+b)
```

```python
#리스트 곱하기(*)
a=[1,2,3]
print(a*3)
```

```python
#리스트의 수정과 삭제
a=[1,2,3]
a[2]=4
print(a)
```

```python
#del로 리스트 요소 삭제하기
a=[1,2,3]
del a[1]
print(a)
```

```python
a=[1,2,3,4,5]
del a[2:]
print(a)
```

```python
#리스트 관련 함수
a=[1,2,3,2,4]
a.append(5) #리스트에 요소 추가하기
print(a)
a.append([6,7]) #리스트에 리스트 추가하기
print(a)
a[6].append(8) #리스트 안에 리스트에 요소 추가하기
print(a)
```

```python
a=[2,1,3,5,4]
a.sort()
print(a) #리스트 정렬하기
a=[2,1,3,5,4]
a.reverse() #리스트 뒤집기
print(a)

```

```python
a=[2,1,3,5,4]	
a.sort(reverse=True) #리스트 정렬하기(내림차순)
print(a)
print(a.index(3))#리스트에서 값이 3인 요소의 인덱스 반환하기
```

```python
a=[1,2,3,4,5]
a.insert(0,10) #리스트의 0번째 위치에 10을 삽입하기
print(a)
```

```python
a=[1,2,3,1,2,3]
a.remove(3) #리스트에서 값이 3인 요소 삭제하기 ->앞에 있는 3만 지워짐.
print(a)
a.remove(3)
print(a)
```

```python
#리스트 요소 꺼내기-pop()
a=[1,2,3,4,5]
print(a.pop()) #리스트의 마지막(기본값) 요소를 꺼내기
print(a)
print(a.pop(1)) #리스트의 1번째 요소를 꺼내기
print(a)
```

```python
#리스트에 포함된 요소 x의 개수 세기 - count()
a=[1,2,3,1,1,3,1,3,1,1,3]
print(a.count(1)) #리스트에 값이 1인 요소가 몇 개인지 세기
print(a.count(2)) #리스트에 값이 2인 요소가 몇 개인지 세기
```

```python
#리스트 확장 - extend()
a=[1,2,3]
b=[4,5,6]
a.extend(b)
print(a)
```
