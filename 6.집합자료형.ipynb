```python
#집합자료형의 특징
##1. 중복을 허용하지 않음.
##2. 순서가 없다.
```

```python
#교집합 구하기
s1=set([1,2,3,4,5,6])
s2=set([4,5,6,7,8,9])

print(s1)
print(s2)

print(s1.intersection(s2)) #intersection 활용
print(s1 & s2) # & 엔퍼센트 활용
```

```python
#합집합 구하기
s1=set([1,2,3,4,5,6])
s2=set([4,5,6,7,8,9])

print(s1.union(s2)) #union 활용
print(s1|s2) #| 파이프 활용
```

```python
#차집합 구하기 difference()
print(s1.difference(s2)) #s1에서의 차집합
print(s2.difference(s1)) #s2에서의 차집합
#- 마이너스 활용
print(s1-s2)
print(s2-s1)
```

```python
#집합자료형 관련 함수
#값 1개 추가하기 - add

s1={1,2,3}
s1.add(4)
print(s1)

#값 여러개 추가하기 -update
s1.update([5,6,7])
print(s1)
print(type(s1))
```

```python
#특정 값 제거하기 -remove
s1={1,2,3}
s1.remove(2)
print(s1)
```

```python
#모든값 제거하기 -clear
s1={1,2,3}
s1.clear()
print(s1)
```
