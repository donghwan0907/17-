```python
for i in range(1,30):
    print(('Fizz'*(i%3==0)+'Buzz'*(i%5==0))or i,end=" ")

#(i%3==0), (i%5==0)은 bool타입이다.
#print("Fizz"*True)  는 출력이 되고
#print("Fizz"*False)  는 출력이 되지 않는다.
```

```python
#심사문제
a,b=map(int,input().split())
for i in range(a,b+1):
    if i%5==0 and i%7==0:
        print("FizzBuzz")
    elif i%5==0:
        print("Fizz")
    elif i%7==0:
        print("Buzz")
    else:
        print(i)
```
