```python
#재귀호출에 종료 조건 만들기
def hello(count):
    #count가 0이 되면 다시 함수를 호출하지 않고 끝냄.
    if count==0:
        return
    print("Hello World",count)
    count-=1
    hello(count)
```

```python
hello(5)
```

```python
#재귀호출로 팩토리얼 구하기
def fac(n):
    if n==1:
        return 1
    return n*fac(n-1)

print(fac(5))
```
