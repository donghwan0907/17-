```python
#1 a,b를 입력받아 a부터 b까지의 홀수만 출력
a,b=map(int,input().split())
for i in range(a,b+1):
    if i%2==0:
        continue
    print(i,end=" ")
```

```python
#2 한 개의 정수 n을 입력받아 n × n 크기의 별을 출력
n=int(input())
for i in range(n):
    for i in range(n):
        print("*",end="")
    print()
```

```python
#3 줄 수 n을 입력받아 피라미드를 출력
n=int(input())
for i in range(1,n+1):
    print(" "*(n-i+1)+"*"*(2*i-1))
```

```python
#4 a, b를 입력받아 a부터 b까지 3의배수 제외 출력
a,b=map(int,input().split())
for i in range(a,b+1):
    if i%3==0:
        continue
    print(i,end=" ")
```

```python
#5 학생 5명의 점수를 입력받아 가장 높은 점수를 출력
score=list(map(int,input().split()))
big=score[0]

for i in score:
    if i >big:
        big=i
print(big)
```

```python
#6 평균을 ABCF로 나눠서 출력
국어,영어,수학=map(int,input().split())
av=(국어+영어+수학)/3
if av>=90:
    print("A")
elif av>=80:
    print("B")
elif av>=70:
    print("C")
else:
    print("F")
```

```python
#7 문자열을 입력받아 거꾸로 출력
a=input()
print(a[::-1])


for i in reversed(a):
    print(i, end="")
```

```python
#8 정수를 공백으로 입력받아 짝수가 몇 개인지 출력
a=list(map(int,input().split()))
cnt=0
for i in a:
    if i%2==0:
        cnt+=1

print(cnt)
```

```python
#10 정답을 7로 정해두고 사용자가 맞출 때까지 계속 입력
y=7
a=int(input())
while a!=y:
    print("다시 입력")
    a=int(input())

print("정답")
```
