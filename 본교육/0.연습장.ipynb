```python
#1 a,b를 입력받아 a부터 b까지의 홀수만 출력
a,b=map(int,input().split())
for i in range(a,b+1):
    if i%2==0:
        continue
    print(i,end=" ")
```

```python
#2 한 개의 정수 n을 입력받아 n × n 크기의 별을 출력
n=int(input())
for i in range(n):
    for i in range(n):
        print("*",end="")
    print()
```

```python
#3 줄 수 n을 입력받아 피라미드를 출력
n=int(input())
for i in range(1,n+1):
    print(" "*(n-i+1)+"*"*(2*i-1))
```

```python
#33
def 젤긴거찾기(*문자열들):
    젤긴거=문자열들[0]
    for i in 문자열들:
        if len(i)>len(젤긴거):
            젤긴거=i
    return 젤긴거
print(젤긴거찾기("cat","elephant","dog","python"))
```

```python
#35
def profile(name, age=20, **kwargs):
    print(f"이름 : {name}")
    print(f"나이 : {age}")

    for k, v in kwargs.items():
        print(f"{k} : {v}")

profile("홍길동", city="서울", hobby="게임")
```
