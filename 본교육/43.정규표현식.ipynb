```python
#문자열
import re

re.match("hello","hello,world")
```

```python
re.search("hello","hello,world")
re.search("world!$","hello, world!")
```

```python
#지정된 문자열이 하나라도 포함되어 있는지 판단하기
re.match("hello|world", "hello")
```

```python
#범위판단하기
print(re.match("[0-9]*", "1234"))
print(re.match("[0-9]+", "1234"))
print(re.match("[0-9]*", "abcd"))
print(re.match("[0-9]+", "abcd"))
```

```python
print(re.match('a*b','b'))
print(re.match('a+b','b'))
print(re.match('a*b','aab'))
print(re.match('a+b','aab'))
```

```python
#문자가 한 개만 있는지 판단하기
print(re.match('abc?d','abd')) #abd에서 c위치에 c가 0개 있으므로 패턴에 매칭됨.
print(re.match('ab[0-9]?c','ab3c')) #[0-9] 위치에 숫자가 1개 있으므로 패턴에 매칭됨.
print(re.match('ab.d','abxd')) #.이 있는 위치에 문자가 1개 있으므로 패턴에 매칭됨.
```

```python
#문자 개수 판단하기
print(re.match('h{3}','hhhello'))# h가 3개 있는지 판단
print(re.match('(hello){3}','hellohellohelloworld'))# hello가 3개 있는지 판단
```

```python
#전화번호 형식
print(re.match('[0-9]{3}-[0-9]{4}-[0-9]{4}', '010-1000-1000')) #전화번호 형식에 맞는지 판단
print(re.match('[0-9]{2,3}-[0-9]{3,4}-[0-9]{4}', '02-100-1000')) #일반전화번호 형식에 맞는지 판단
```

```python
#숫자와 영문 문자를 조합해서 판단하기
print(re.match('[a-zA-Z0-9]+', 'Hello1234'))#a부터z, A부터 Z, 0부터9까지 1개 이상
print(re.match('[A-Z0-9]+','hello'))#소문자만 있으므로 매칭X
```

```python
#한글은 가-힣 이다.
print(re.match('[가-힣]+','랊끓낊꺏햟'))
print(re.match('[ 가-힣]+','랊 끓 낊 꺏 햟')) #공백을 찾으려면 가 앞에 공백을 입력한다. (특수기호도 포함)
print(re.match('[!@#$% 가-힣]+','랊 끓 낊$ %#꺏! 햟'))
```

```python
#특정 문자 범위에 포함되지 않는지 판단하기
print(re.match('[^A-Z]+', 'hello1234#$'))#^는 해당하는 대문자를 제외 모든 문자가 있는지 판단
```

```python
#모든 범위 지정방법
print(re.match(r'\d+', '1234')) #모든 숫자 : [0-9]와 같음
print(re.match(r'\D+', 'Hello'))#숫자 제외 모든 문자 : [^0-9]와 같음
print(re.match(r'\w+', 'Hello_1234'))#영문 대소문자, 숫자, 밑줄 문자 : [A-Za-z0-9]와 같음
print(re.match(r'\W+', '(:)'))#영문 대소문자, 숫자, 밑줄 문자 제외 모든 문자 : [^a-zA-Z0-9]와 같다고 하는데 다르네
```

```python
print(re.match('[^a-zA-Z0-9]+', '(:응)'))
```

```python
#그룹 사용하기
m = re.match('([0-9]+) ([0-9]+)', '10 295')
print(m.group(1))
print(m.group(2))
print(m.group())
print(m.group(0))
print(m.groups())
```

```python
# 이름정하기 = ?P<이름>정규표현식
m = re.match(r'(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\((?P<arg>\w+)\)','print(1234)')
print(m.group('func'))
print(m.group('arg'))
```

```python
#패턴에 매칭되는 모든 문자열 가져오기
re.findall('[0-9]+', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
#문자열 바꾸기
# re.sub("패턴","바꿀문자열","문자열",바꿀횟수)
re.sub('apple|orange','fruit','apple box orange tree')
```

```python
re.sub('[0-9]+','n', '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
#교체함수 -> 바꿀문자열 자리에 함수 넣기
def multiple10(m):
    n=int(m.group())
    return str(n*10)

re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
#람다함수도 가능.
re.sub('[0-9]+',lambda m: str(int(m.group())*10), '1 2 Fizz 4 Buzz Fizz 7 8')
```
