```python
#import로 모듈 가져오기
import math as m

print(m.pi) #파이 구하기
print(m.sqrt(4.0)) #제곱근 구하기
```

```python
#모듈 일부만 가져오기
from math import pi
print(pi)
```

```python
#모듈의 모든 것을 가져오기
from math import *
print(pi)
print(sqrt(4))
```

```python
#모듈 일부를 가져오고 이름 지정하기
from math import pi as 파이, sqrt as 제곱근
print(제곱근(81))
print(파이)
```
