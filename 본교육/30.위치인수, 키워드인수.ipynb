```python
#위치 인수를 사용하는 함수
def p_num(a,b,c):
    print(a)
    print(b)
    print(c)
```

```python
p_num(10,20,30)
```

```python
#언패킹 사용하기
x=[40,50,60]
print(x)
print("ㅡㅡ")
print(*x)
print("ㅡㅡ")
p_num(*x)
```

```python
#가변인수함수
def p_num(*args):
    for arg in args:
        print(arg)
```

```python
p_num(10)
p_num(10,20,30,40)
```

```python
x=[10]
p_num(*x)
p_num(x)
```

```python
y=[10,20,30,40]
p_num(*y)
p_num(y)
```

```python
#키워드 인수 사용하기
def perinfo(name,age,address):
    print('이름:',name)
    print('나이:',age)
    print('주소:',address)

perinfo("홍길동","30","서울시 용산구")
```

```python
#키워드인수로 사용하면 순서를 이름 나이 주소가 아닌 순서로 써도 출력됨.
perinfo(name='홍길동',address="서울시 용산구",age=30)
```

```python
#키워드인수와 딕셔너리 언패킹
def perinfo(name,age,address):
    print('이름:',name)
    print('나이:',age)
    print('주소:',address)
    
x= {'name':'홍길동','age':30,'address':'서울시 용산구'}
perinfo(**x)
#별표를 두 번 넣어야 함(한번 넣으면 키가 출력, 두번 넣어야 값이 출력.)
```

```python
#키워드 인수를 사용하는 가변인수함수 만들기 (**)
def perinfo(**kwargs):
    for kw, arg in kwargs.items():
        print(kw,':', arg,sep='')

perinfo(name='홍길동',age=30,address='서울시 용산구')
```

```python
y={'name':'홍길동','age':30,'address':'서울시 용산구'}
perinfo(**y)
```

```python
#매개변수에 초깃값 지정하기
def perinfo(name,age,address='비공개'):
    print('이름:',name)
    print('나이:',age)
    print('주소:',address)
```

```python
#심사문제
점수=list(map(int,input().split()))
def min_score(*args):
    return min(args)
def max_score(*args):
    return max(args)
def avg_score(*args):
    return sum(args)/len(args)
def score(*args):
    print(f"낮은 점수: {min_score(*args):.2f}, 높은 점수: {max_score(*args):.2f}, 평균 점수: {avg_score(*args):.2f}")

score(*점수)
a=[점수[1],점수[3]]
score(*a)
```
