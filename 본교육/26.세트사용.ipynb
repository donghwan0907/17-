```python
#집합연산 사용
a={1,2,3,4}
b={3,4,5,6}
#합집합
print(a|b) 
print(a.union(b))
print(set.union(a,b))
 #교집합
print(a&b)
print(a.intersection(b))
print(set.intersection(a,b))
#차집합
print(a-b)
print(a.difference(b))
print(set.difference(a,b))
#대칭차집합
print(a^b)
print(a.symmetric_difference(b))
print(set.symmetric_difference(a,b))
```

```python
#~세트 조작하기까지 건너뜀.
#세트 조작하기.
#세트 추가하기
a={1,2,3,4}
a.add(5)
print(a)
#세트 삭제하기
a.remove(3)
print(a)
a.discard(2)
print(a)
```

```python
a={1,2,3,4}
a.pop()
print(a)
#len 도 가능
print(len(a))
a.clear()
print(a)
```

```python
#반복문으로 세트 요소 출력하기
a={1,2,3,4}
for i in a:
    print(i)
```

```python
#세트 표현식 사용
#{식 for 변수 in 반복가능한 객체}
#set{식 for 변수 in 반복가능한 객체}

a={j for j in "apple"}
print(a)
```

```python
#세트 표현식에 if 조건문 사용하기
#{식 for in 세트 if조건식}
#set(식 for in 세트 if조건식)

a={i for i in 'pineapple' if i not in "apl"}
print(a)
```

```python
#심사문제
a,b=map(int,input().split(" "))
c=max(a,b)
for i in range(1,c+1):
    if i % a==0 and i%b==0:
        print(i,end=" ")
print()
```
