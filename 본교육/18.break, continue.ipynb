```python
for i in range(10000):
    print(i)
    if i == 10:
        break
```

```python
#continue로 코드 실행 건너뛰기 (홀수만 출력하기)
for i in range(20):
    if i%2==0:
        continue
    print(i)
```

```python
i=0
while i<20:
    i+=1
    if i%2==0:
        continue
    print(i)
```

```python
i=0
while True:
    if i %10 !=3:
        i+=1
        continue
    if i>73:
        break
    print(i,end=" ")
    i +=1
```

```python
a, b = map(int, input().split())
for i in range(a, b + 1):
    if i % 10 == 3:
        continue
    print(i, end=" ")
```

```python
#심사문제
a, b = map(int, input().split())
while True:
    if a == b+1:
        break

    if a % 10 == 3:
        a += 1
        continue

    print(a)
    a += 1
```
