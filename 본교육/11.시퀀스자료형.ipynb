```python
#시퀀스자료형의 공통기능
#시퀀스자료형 : 리스트 튜플 랜지
a=list(range(0,91,10))
print(type(a))
print(a)
print(30 in a)
print(100 in a)
```

```python
#시퀀스 객체 연결하기
a=[0,10,20,30]
b=[9,8,7,6]

print(a+b)
```

```python
#range(10)+range(10,20) range는 +로 객체를 연갈할 수 없다. 리스트로 만들고 연결하여야 한다.
list(range(10))+list(range(10,20))
```

```python
#시퀀스 객체 반복하기
a=list(range(0,31,10))
print(a*3)
```

```python
#시퀀스 객체의 요소 개수 구하기 len
a=list(range(0,10))
print(a)
print(len(a))
```

```python
#range의 숫자 생성 갯수 구하기
c=tuple(range(1,546123,112587))
print(len(c))
print(c)
```

```python
#문자열의 길이 구하기
hi="hello world 안녕"
print(len(hi))
```

```python
#인덱스 사용하기
a=list(range(1,120,6))
print(a)
print(a[2])
print(a[-1])
print(a[0])

b="hello world!"
print(b)
print(b[0])
print(b[-1])
print(b[6])
r=len(b)
print(r)
print(b[r-1])
```

```python
#요소에 값 할당하기 (튜플은 수정 불가)
a=[0,0,0,0,0000000]
a[0]=38
a[2]=21
print(a)
```

```python
#요소 삭제하기
a=[1,2,3,4,5,6,7,8,9]
del a[2]
print(a)
```

```python
#슬라이싱하기
a=list(range(0,10))
b=a[-9:4]
c=a[4:-1]
print(a)
print(b+c)
print(a[::2]) #증가폭 2로 설정
```

```python
a=list(range(0,100,10))
print(a[:])
print(a[::-1])
```

```python
#len 응용하여 슬라이싱하기
a=list(range(0,100,10))
print(a[:len(a)])
print(a[5:len(a)])
```

```python
#문자열에 슬라이스 사용하기
a="hello  my name is donghwan jeong."
print(a[:5])
print(a[::2])
print(a[::-1])
```

```python
#슬라이스에 요소 할당
a=list(range(0,100,10))
b=list(range(0,10))
a[2:5]=['a','b','c']
a[7:9]=b[7:9]
print(a)
```

```python
a=list(range(0,100,10)) #인덱싱보다 할당할 요소가 많으면 그만큼 추가 가능
a[1:2]=['a','b','c','d']
print(a)
```
