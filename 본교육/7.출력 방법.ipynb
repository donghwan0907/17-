```python
#값을 여러개 출력하기
print(1,2,3)
print("hello", "python")
```

```python
#sep로 값 사이에 문자 넣기
print(1,2,3,sep=",") #값 사이에 ,를 넣어라
print(1,2,3,sep="") #값 사이에 공백을 넣어라
print(1920,1000,sep="X")
```

```python
#줄바꿈 활용
print(1,2,3, sep='\n')
print("dafds\nasdfasdf\nasdf")
```

```python
print('1\n2\n3')
```

```python
#end 활용
print(1, end="")
print(2, end="")
print(3, end="")
```

```python
year=2000
month=10
day=27
hour=11
minute=43
second=59

print(year,month,day,sep='/',end=' ')
print(hour,minute,second,sep=':')
```

```python
#람다표현식 자체를 호출하기
#(lambda 매개변수들 : 식)(인수들)
(lambda x:x+10)(2)
```
