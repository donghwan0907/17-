```python
#나눗셈 후 소수점 이하를 버리는 연산자 //
print(5/3)
print(5//3)
```

```python
#나눗셈 후 나머지를 구하는 연산자 %
print(5%3)
```

```python
#거듭제곱을 구하는 **연산자
print(2**10)
```

```python
#값을 정수로 만들기
# int(숫자) int(계산식) int('문자열')
a=int(3.3)
b=int(5/2)
c=int('10')

print(a);print(type(a))
print(b);print(type(b))
print(c);print(type(c))
```

```python
# 값을 실수로 만들기
a=float(5)
b=float(5.521)
c=float('5.3')

print(a);print(type(a))
print(b);print(type(b))
print(c);print(type(c))
```
