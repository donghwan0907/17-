```python
def add(a,b):
    """이 함수는  a와b를 더한 뒤 결과를 반환하는 함수입니다."""
    return a+b
x=add(10,20)
print(x)
print(add.__doc__)
```

```python
#함수에서 값을 여러 개 반환하기
def add_sub(a,b):
    return a+b, a-b

x=add_sub(10,20)
print(x)
x,y=add_sub(10,20)
print(x,y)
```

```python
#함수의 호출 과정 알아보기
def mul(a,b):
    c=a*b
    return c
def add(a,b):
    c=a+b
    print("합:",c)
    d=mul(a,b)
    print("곱:",d)
x=10
y=20
add(x,y)
```

```python
#심사문제
a,b=(map(int,input().split()))
def 덧뺄곱나(a,b):
    return a+b,a-b,a*b,a/b
x,y,z,r=덧뺄곱나(a,b)
print(f"덧셈: {x},뺼셈: {y}, 곱셈: {z}, 나눗셈: {r}")
```
