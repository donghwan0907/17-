```python
#반복문으로 2차원 리스트 모두 출력
a=[[10,20],[30,40],[50,60]]
for i in a:
    print(i)

for x,y in a:
    print(x,y)
```

```python
#for 반복문 두번사용하기
a=[[10,20],[30,40],[50,60]]
for i in a: #a의 3개 요소가 i
    for j in i: #a의 3개 요소마다 있는 2개 요소들이 j
        print(j,end="  ")
    print()
```

```python
#인덱스를 반복하여 프린트하기
a=[[10,20],[30,40],[50,60]]
for i in range(len(a)):
    for j in range(len(a[i])):
        print(a[i][j],end=" ")
    print()
```

```python
#while반복문을 한 번 사용하기
a=[[10,20],[30,40],[50,60]]
i=0
while i<len(a):
    x,y=a[i]
    print(x,y)
    i+=1
```

```python
#while 두번사용하기
a=[[10,20],[30,40],[50,60]]
i=0
while i<len(a):
    j=0
    while j<len(a[i]):
        print(a[i][j],end=" ")
        j+=1
    print()
    i+=1
```

```python
# 가로(행), 세로(열) 입력
가로, 세로 = map(int, input().split())

# 지도 입력
board = []
for i in range(가로):
    board.append(list(input()))

# 8방향
dx = [-1,-1,-1,0,0,1,1,1]
dy = [-1,0,1,-1,1,-1,0,1]

for i in range(가로):
    for j in range(세로):

        # 지뢰면 그대로 출력
        if board[i][j] == '*':
            print('*', end='')
            continue

        count = 0

        # 주변 8칸 검사
        for k in range(8):
            nx = i + dx[k]
            ny = j + dy[k]

            # 범위 안인지 확인
            if 0 <= nx < 가로 and 0 <= ny < 세로:
                if board[nx][ny] == '*':
                    count += 1

        print(count, end='')

    print()
```

```python
row,col=map(int,input().split())
matrix=[]
for i in range(row):
    matrix.append(list(input()))
for r in range(row):
    for c in range(col):

        if matrix[r][c]=="*":
            continue

        count=0
        for nr in range(r-1,r+2):
            for nc in range(c-1,c+2):
                if 0<=nr<row and 0<=nc<col:
                    if matrix[nr][nc]=="*":
                        count+=1
        matrix[r][c]=str(count)
for r in range(row):
    print("".join(matrix[r]))
```
