```python
#딕셔너리 샤용 (키:밸류)
lux={'health':490,'mana':334,'melee':550,'armor':18.72}
print(lux)
```

```python
과일가격 = {"사과": "과일","바나나": 2000,"포도": 2000,"멜론": 2000}
print("사과" in 과일가격)#키값만 찾을 수 있다.
print("과일" in 과일가격)
```

```python
##딕셔너리에서 벨류값으로 키 찾기
#1 중복되는 벨류값의 모든 키 출력(반복)
과일가격 = {"사과": 2000,"바나나": 2000,"포도": 2000,"멜론": 2000}
for k, v in 과일가격.items():
    if v == 2000:
        print(k)
#2 첫번째 키만 출력
print(list(과일가격.keys())[list(과일가격.values()).index(2000)])
```

```python
#딕셔너리 개수
len(과일가격)
```

```python
과일가격 = {"사과": 1000,"바나나": 2000,"포도": 3000}
#딕셔너리 추가하기
과일가격['귤']=5000
print(과일가격)
```

```python
과일가격 = {"사과": 1000,"바나나": 2000,"포도": 3000}
#딕셔너리 키에 값 변경하기
과일가격['바나나']=15000
print(과일가격)
```

```python
#pop으로 딕셔너리 꺼내기
과일가격 = {"사과": 1000,"바나나": 2000,"포도": 3000}
a=과일가격.pop("바나나")
print(과일가격)
print(a) #pop으로 꺼낸 항목을 다른 변수에 할당 가능. del은 삭제만 가능하여 할당 불가.
```

```python
#키값 바꾸기
과일가격 = {"사과": 1000,"바나나": 2000,"포도": 3000}
과일가격["귤"] = 과일가격.pop("바나나")
```
