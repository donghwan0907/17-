```python
#함수 안에서 전역변수 변경하기
b=10
def foo():
    b=20
    print(b)
foo() #지역변수 b=20 출력
print(b) #전역변수 b=10에 영향을 주지 않아 그대로 출력
```

```python
c=10
def foo():
    global c #global 은 지역변수로만 쓰지 않고 전역변수로 업데이트를 하겠다는 명령어.
    c=20
    print(c)
foo()
print(c)
```

```python
def asdf():
    global dd
    dd=50
    print(dd)
asdf()
print(dd)
```

```python
#함수 안에서 함수 만들기
#def 함수이름1():
#    코드
#    def 함수이름2():
#        코드

def print_hello():
    hello='Hello, World!'
    def print_message():
        print(hello)
    print_message()

        
print_hello()
```

```python
#지역변수 변경하기
def A():
    x=10
    def B():
        nonlocal x #제일 가까운 x를 불러오기(A함수의 지역변수인 x)
        x=20
    B()
    print(x)
A()
```

```python
#논로컬이 변수를 찾는 순서
def A():
    x=10
    y=100
    def B():
        x=20
        def C():
            nonlocal x
            nonlocal y
            x=x+30
            y=y+300
            print(x)
            print(y)
        C()
    B()
A()
```

```python
#glovbal로 전역 변수 사용하기
x=1
def A():
    def B():
        x=20
        def C():
            global x
            x+=30
            print(x)
        C()
    B()
A()
```

```python
#클로저 사용하기
def calc():
    a=3
    b=5
    def mul_add(x):
        return a*x+b
    return mul_add
c=calc()
print(c(1),c(2),c(3),c(4),c(5))
```

```python
#lambda로 클로저 만들기
def calc():
    a=3
    b=5
    return lambda x:a*x+b
c=calc()
print(c(1),c(2),c(3),c(4),c(5))
```

```python
#클로저의 지역 변수 변경하기
def calc():
    a=3
    b=5
    total=0
    def mul_add(x):
        nonlocal total
        total=total+a*x+b
        print(total)
    return mul_add
c=calc()
c(1)
c(2)
c(3)
```
