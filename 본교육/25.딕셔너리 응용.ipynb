```python
#딕셔너리 조작하기
#setdefault, update로 키, 벨류 추가하기

a={
    "a":10,
    "b":20,
    "c":30,
    "d":40
}

a.setdefault("e")
print(a)
```

```python
a.setdefault("f",100)
print(a)
```

```python
#딕셔너리의 키의 값 수정
a={
    "a":10,
    "b":20,
    "c":30,
    "d":40
}

a.update(a=90)
print(a)
```

```python
#e라는 키가 없으면 추가됨.
a.update(e=90)
print(a)
```

```python
a.update(a=900,f=6)
print(a)
```

```python
y={
    1:"one",
    2:"two"
}
print(y)
y.update(
    {1:"ONE",2:"TWO"}
)
print(y)
```

```python
y.update(
    [
    [2,"two"],
    [4,"four"]
    ]
)
print(y)
```

```python
y.update(
    zip([1,2],["one","two"])
)
print(y)
```

```python
#딕셔너리 표현식 사용 (컴프리헨션)
keys=['a','b','c','d']
x={key: value for key, value in dict.fromkeys(keys).items()}
x
```

```python
x={"a":10, "b": 20,"c":30, "d": 40}
for key, value in x.items():
    print(key, value)
```

```python
#딕셔너리 표현식에서 if 사용하기
x={"a":10, "b": 20,"c":30, "d": 40}
for key, value in x.items():
    if value==20:
        del x[key]
print(x)

#중간에 값이 20인 키 b가 지워지면서, 딕셔너리의 크기가 바뀌는데 여기서 에러가 발생한다.
#그래서 다음과 같이 딕셔너리를 새로 생셩해야한다.
```

```python
x={"a":10, "b": 20,"c":30, "d": 40, "e":20,}
y={key: value for key, value in x.items() if value !=20}
print(y)
print(x)
```
