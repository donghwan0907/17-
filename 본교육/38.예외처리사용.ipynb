```python
#특정 예외만 처리하기
'''
try:
    실행할 코드
except 예외이름:
    예외가 발생했을 때 처리하는 코드
'''

y=[10,20,30]
try:
    index, x=map(int,input('인덱스와 나눌 숫자를 입력하세요:').split())
    print(y[index]/x)
except ZeroDivisionError:
    print('숫자를 0으로 나눌 수 없습니다.')
except IndexError:
    print('잘못된 인덱스입니다.')
```

```python
#예외의 에러 메시지 받아오기
'''
try:
    실행할 코드
except 예외 as 변수:
    예외가 발생했을 때 처리하는 코드
'''
y=[10,20,30]
try:
    index, x=map(int,input('인덱스와 나눌 숫자를 입력하세요:').split())
    print(y[index]/x)
except ZeroDivisionError as e:
    print('숫자를 0으로 나눌 수 없습니다.',e)
except IndexError as e:
    print('잘못된 인덱스입니다.',e)
```

```python
#else finally 사용
'''
try:
    실행할 코드
except:
    예외가 발생했을 때 처리하는 코드
else:
    예외가 발생하지 않았을 때 실행할 코드
'''
try:
    x=int(input('나눌 숫자를 입력하세요:'))
    y=10/x
except ZeroDivisionError:
    print("숫자를 0으로 나눌 수 없습니다.")
else:
    print(y)
```

```python
#예외와는 상관없이 항상 코드 실행
try:
    x=int(input('나눌 숫자를 입력하세요:'))
    y=10/x
except ZeroDivisionError:
    print("숫자를 0으로 나눌 수 없습니다.")
else:
    print(y)
finally:
    print("코드 실행 끝")
```

```python
#예외 발생시키기
try:
    x=int(input('3의 배수 입력:'))
    if x%3 !=0:
        raise Exception('3의 배수가 아닙니다.')
    print(x)
except Exception as e:
    print('예외발생 :',e)
```

```python
#현재 예외를 다시 발생시키기
def three_multiple():
    try:
        x=int(input('3의 배수를 입력하세요:'))
        if x%3 !=0:
            raise Exception('3의 배수가 아닙니다.')
        print(x)
    except Exception as e:
        print('three_multiple 함수에서 예외가 발생했습니다.',e)
        raise
try:
    three_multiple()
except Exception as e:
    print('스크립트 파일에서 예외가 발생했습니다.',e)
```
