```python
#1
students = {
    "김철수": [90, 80, 70],
    "이영희": [100, 95, 90],
    "박민수": [60, 70, 65]
}

print("===== 학생 성적 결과 =====")

total = 0

for name, score in students.items():
    avg = sum(score) / len(score)
    total += avg

    print()
    print("이름:", name)
    print("점수:", score)
    print("평균:", avg)

print()
print("===== 전체 결과 =====")
print("전체 학생 수:", len(students), "명")
print("전체 평균:", total / len(students))
```

```python
#5
students = []

for i in range(5):
    print(i + 1, "번째 학생")

    name = input("이름: ")
    age = int(input("나이: "))
    grade = input("학년: ")

    student = {
        "이름": name,
        "나이": age,
        "학년": grade
    }

    students.append(student)

print()

for s in students:
    print(s)
```

```python
#8
a,b=map(int,input().split())

for i in range(a,b-1,-1):
    if i%3==0 and i%7==0:
        print("FizzBuzz",end=(''))
    elif i%3==0:
        print("Fizz",end=(''))
    elif i%7==0:
        print("Buzz",end=(''))
    else:
        print(i, end=(' '))
print()
```

```python
n = int(input())

# 위쪽
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))

# 아래쪽
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
    
```
