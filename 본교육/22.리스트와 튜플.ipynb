```python
#리스트에 요소 추가 append, extend. insert
a=list(range(10,31,10))
a.append(40)
print(a)
```

```python
#리스트 안에 리스트 추가하기
a=[10,20,30]
a.append([500,600])
print(a)
```

```python
#리스트 확장하기
a=[10,20,30]
a.extend([500,600])
print(a)
```

```python
#리스트의 특정 인덱스에 요소 추가하기
a=[10,20,30]
a.insert(0,500)
print(a)
```

```python
a=[10,20,30]
a.insert(len(a),500)
print(a)
```

```python
#리스트 수정하기 
a=[10,20,30]
a[1:3]=[500,600]
print(a)
```

```python
#[1:1]은 인덱스 존재하지 않지만, 그 부분에 추가 가능
a=[10,20,30]
a[1:1]=[500,600]
print(a)
```

```python
#list 요소 삭제하기
#pop으로 꺼내어 리스트에서 삭제가능, 꺼낸 요소 할당 가능.
a=[10,20,30]
b=a.pop(0)
print(a)
print(b)
```

```python
#del로 삭제 가능, 팝과 달리 꺼낸 요소 할당 불가
a=[10,20,30]
del a[0]
print(a)
```

```python
#remove로 인덱스가 아닌 값 지정하여 삭제하기
a=[10,20,30]
a.remove(20)
print(a)
```

```python
a=[10,20,30,20]
a.remove(20)
print(a)
#앞에 20만 지워짐
```

```python
#리스트에서 특정 값의 인덱스 구하기
a=[10,20,30,15,20,40]
a.index(20)
#처음 20의 인덱스를 구함.
```

```python
#특정 값의 개수 구하기
a=[10,20,30,15,20,40]
a.count(20)
```

```python
#list 순서 뒤집기
a=[10,20,30,15,20,40]
a.reverse()
print(a)
```

```python
#list요소 정렬하기
a=[10,20,30,15,20,40]
a.sort()
print(a)
```

```python
#list 모든 요소 삭제
a=[10,20,30,15,20,40]
a.clear()
print(a)
```

```python
#list를 슬라이스로 조작하기
a=[10,20,30,15,20,40]
a[len(a):]=[500]
a
```

```python
#list 할당과 복사
a=[0,0,0,0,0]
b=a
b[2]=99
print(a)
a is b
```

```python
#a를 b에 복사하기.
a=[0,0,0,0,0]
b=a.copy()
a is b
```

```python
#인덱스와 요소를 함께 출력하기
a=[38,21,53,62,19]
for i,v in enumerate(a):
    print(f"{i+1}번째 값은 {v}입니다.")
```

```python
a=[38,21,53,62,19]
i=0
while i <len(a):
    print(a[i])
    i+=1
```

```python
#list의 가장 작은수, 큰수 구하기
a=[38,21,53,62,19]
sm=a[0]
big=a[0]
for i in a:
    if i < sm:
        sm=i
    if i > big:
        big=i
print(sm,big)
```

```python
#함수를 사용하여 작은수, 큰수, 합계 구하기
a=[38,21,53,62,19]
print(max(a))
print(min(a))
print(sum(a))
```

```python
#리스트 map 사용
a=list(map(int,input().split()))
print(a)
```

```python
a=list(map(str,range(10)))
print(a)
print(type(a[0]))
b=list(range(10))
print(b)
print(type(b[0]))
```

```python
#튜플에서 특정값의 인덱스, 개수 구하기
a=(38,21,53,62,19,53)
print(a.index(53))
print(a.count(53))
```

```python
#리스트에서 사용했던 함수 map min max sum 튜플에서도 사용가능.
a=tuple(map(int,input().split()))
print(a)
print(sum(a))
print(max(a))
print(min(a))
print(type(a))
```

```python
#심사문제
a,b=map(int,input().split())
c=[]
for i in range(a,b+1):
    c.append(2**i)
print(c)
```
