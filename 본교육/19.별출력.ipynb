```python
for i in range(5):
    for j in range(5):
        print("j:",j,sep="",end=" ")
    print("i",i,"\\n",sep="")
```

```python
for i in range(5):
    for j in range(5):
        print("*",end=" ")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j<=i:
            print("*",end=" ")
    print()
```

```python
for i in range(5):
    for j in range(5):
        if j==i:
             print("*",end="")
        else:
            print("  ", end="")
    print()
```

```python
#심사문제
n = int(input())

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
#심사문제
h=int(input())
for i in range(h):
    for j in range(h-i-1):
        print(" ",end="")
    for k in range(2*i+1):
        print("*",end="")
    print()
```
