```python
#for 반복문
#for 변수 in range(횟수):
#    반복할 코드
```

```python
for i in range(10):
    print("hello",i)
```

```python
for i in range(5,12,2):
    print("hello",i)
```

```python
for i in reversed(range(10)):
    print("hi",i)
```

```python
for i in range(10,0,-1):
    print("hi",i)
```

```python
#입력횟수대로 반복
cnt=int(input())
for i in range(cnt):
    print('hi',i)
```

```python
cnt=int(input())
for i in range(cnt,0,-1):
    print('hello',i,cnt)
```

```python
#시퀀스 객체로 반복
a=list(range(10,51,10))
for i in a:
    print(i)
```

```python
#튜플
fruits=("apple","orange","banana")
for fruit in fruits:
    print(fruit)
```

```python
for i in 'python':
    print(i, end=", ")
```

```python
for i in reversed('donghwan'):
    print(i,end=" ")
```

```python
#심사문제
단=int(input())
for i in range(1,10):
    print(단,"*",i,"=",i*단)
```
