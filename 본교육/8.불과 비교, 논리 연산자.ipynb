```python
#비교 연산자의 판단 결과는 True False로 사용한다.
print(3>1)
print(5==5)
print(5==2)
print(5!=5)
print("python"=="python")
print(1==1.0000000000)
print(1 is not 1.0)
```

```python
print(True and True)
print(True and False)
print(True or False)
```

```python
print(not True)
print(not False)
```
