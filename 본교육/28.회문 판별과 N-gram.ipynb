```python
#회문 판별하기
word=input('단어를 입력하세요:')
회문=True
for i in range(len(word)//2):
    if word[i] !=word[-1 -i]:
        회문=False
        break
print(회문)
```

```python
#시퀀스 뒤집기로 문자 검사하기
word=input()
print(word==word[::-1])
```

```python
#리스트와 reversed사용
word="level"
list(word)==list(reversed(word))
```

```python
#문자열의 join메서드와 reversed사용하기
word='level'
word==''.join(reversed(word))
```

```python
#N-gram 만들기
#반복문으로 출력하기
t='Hello'
for i in range(len(t)-1):
    print(t[i],t[i+1],sep='')
```

```python
t='this is python script'
w=t.split()
for i in range(len(w)-1):
               print(w[i],w[i+1])
```

```python
#zip으로 2-gram만들기
t='Hello'
two_gram=zip(t,t[1:])
for i in two_gram:
    print(i[0],i[1],sep='')
```

```python
#zip과 리스트 표현식으로 만들기
t='hello'
[t[i:] for i in range(3)]
```
