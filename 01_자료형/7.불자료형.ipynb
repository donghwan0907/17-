```python
#if 0:
#	   수행할 문장
print(bool("python"))
print(bool(""))
print(bool(" "))
print("------------")
print(bool(10))
print(bool(0)) #0 빼고는 다 true
print("------------")
print(bool([1,2,3]))
print(bool([]))
print("------------")
print(bool({"name":"pey"}))
print(bool({}))

```

```python
#논리연산자

##and 연산자 -> 하나라도 false가 있으면 False이다.
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
#or 연산자 -> True가 하나라도 있으면 True
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True and False)
```

```python
#not 연산자
print(not True)
```
