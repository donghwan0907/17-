```python
#문자열 포매팅
##1. 숫자 바로 대입
"I eat %d apples." % 3
print("I eat %d apples." % 3)
##2. 문자열 바로 대입
"I eat %s apples." % "five"
print("I eat %s apples." % "five")
```

```python
##3. 숫자 값을 나타내는 변수로 대입
number=3
print("I eat %d apples." % number)
##4. 2개 이상의 값 넣기
number=10
day="three"
"I eat %d apples. so I was sick for %s days." % (number,
day)
print("I eat %d apples. so I was sick for %s days." % (number, day))
```

```python
#format 함수를 사용한 포매팅 %d %s 바꿀필요 없음.
print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))
```

```python
#format 함수를 이용해서 2개이상의 값 넣기
number=10
day="three"
print("I eat {0} apples. so I was sick for {1} days.".format(number, day))
```

```python
# f 문자열 포매팅
name="홍길동"
age=10
print("나의 이름은 {name}입니다. 나이는 {age}입니다.")
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.")
```

```python
#문자열 관련 함수들

##문자 개수 세기 - count
a="hobby"
print(a.count('b'))
```

```python
## 위치 알려주기 - find
a="Python is the best choice"
print(a.find('b'))
print(a.find('k')) #찾는 문자열이 존재하지 않으면 -1을 반환한다. find는 왼쪽에서 오른쪽으로 검색한다.
print(a.rfind('i')) #rfind는 오른쪽에서 왼쪽으로 검색한다.
```

```python
#위치 알려주기 - index
a="Life is too short"
print(a.index('t')) #index는 찾는 문자열이 존재하지 않으면 에러를 발생시킨다, 두 개 이상 존재하는 경우 왼쪽에서 오른쪽으로 검색한다.
print(a.rindex('i')) #rindex는 오른쪽에서 왼쪽으로 검색한다.
```

```python
#공백지우기 - strip
a="   hello   "
print(a)
a=a.strip()
print(a)
```

```python
#공백지우기 왼쪽/오른쪽
a="   hello   "
print(a.lstrip())
print(a.rstrip())
```
