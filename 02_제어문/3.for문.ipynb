```python
a=[(1,2),(3,4),(5,6)]
for (t) in a:
	print(t)
print("-----------")
a=[(1,2),(3,4),(5,6)]
for (k,j) in a:
	print(j)
print("-----------")
for (k,j) in a:
	print(k)
```

```python
marks=[90,25,67,45,80]

num=0
for m in marks:
	num =num+1
	if m >=60:
		print(num,"학생은 합격입니다.")
	else:
		print(num,"학생은 불합격입니다.")
```

```python
#for문과 continue
marks=[90,25,67,45,80]
num=0
for m in marks:
	num+=1
	if m <60:
		continue
	print(num,"학생은 합격입니다.")
```

```python
#for와 range를 이용한 구구단
for i in range(2,4):
	for j in range(1,10):
		print(i,"X",j,"=",i*j)
```

```python
#리스트 컴프리헨션
a=[1,2,3,4]
result = []

for num in a:
	result.append(num*3)

print(result)
print(a)
```

```python
#for문과 break문
for i in range(10): #= rnage(0,10) :가 아니라 , 을 써야한다
	if i==5:
		break
	print(i)
```

```python
#enumerate() 함수
fruits=['apple','banana','orange']
for fruit in fruits:
	print(f"과일: {fruit}")
```

```python
#enumerate() 함수
fruits=['apple','banana','orange']
for fruit in enumerate(fruits):
	print(f"과일: {fruit}")
```

```python
#enumerate() 함수
fruits=['apple','banana','orange']
for i,fruit in enumerate(fruits):
	print(f"{i+1}번째 과일: {fruit}")
```

```python
#zip함수로 여러 리스트 순회하기
names=['홍길동','김철수','이영희']
scores=[85,92,78]

for name, score in zip(names,scores):
	print(f"{name}: {score}점")
```
