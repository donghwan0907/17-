```python
treehit=0
while treehit < 10:
	treehit = treehit + 1
	print(f"나무를{treehit}번 찍었습니다")
	if treehit == 10:
		print("나무가 넘어갑니다")
```

```python
count=11
while count !=0:
	count=count-1
	print(count)
	if count==0:
		print("공개")
```

```python
coffee=5
cost=300

while cost: #숫자는 0이 아니면 True
	print("돈을 받았으니 커피를 드립니다.")
	coffee=coffee-1
	print("남은 커피는",coffee,"개 입니다.")
	if coffee ==0:
		print("커피가 다 떨어졌습니다.")
		break
```

```python
coffee =5
while True:
	cost = int(input("돈을 넣어주세요:"))
	if cost ==300:
		print("커피를 드립니다.")
		coffee -=1      #  =  coffee=coffee-1
	elif cost>300:
		print("거스름돈",cost-300,"를 주고 커피를 줍니다.")
		coffee-=1
	else:
		print("돈을 다시 돌려주고 커피를 드리지 않습니다.")
		print("현재 남은 커피는",coffee,"개 입니다.")
	if coffee ==0:
		print("커피가 다 떨어졌습니다.")
		break
```

```python
#while문의 맨 처음으로 돌아가기
a=0
while a <10:
	a=a+1
	if a%2==0:
		continue  #나머지가 0이면 그 다음으로 가지 않고 다시 while로 간다. (프린트를 하지 않는다.)
	print(a)
```
