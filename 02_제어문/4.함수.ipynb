```python
#일반적인 함수
##매개변수 0, 반환값 0

#def 함수이름(매개변수):
#	수행할문장1
#	수행할문장2
#	...
#	return 반환값
```

```python
def add(a,b):
	result=a+b
	return result

a=add(3,4)
print(a)
```

```python
#입력값이 없는 함수
##매개변수 없고 반환값은 있음

def say():
	return 'HI'
a=say()
print(a)
```

```python
#반환값이 없는 함수
##매개변수 있고 반환값 없음

def add(a,b):
	print(f"{a},{b}의 합은{a+b}")
	#(return None)
a=add(3,4)
print(a)
```

```python
#입력값, 반환값 모두 없는 함수
#매개변수 반환값 모두 없음

def say():
	print("hi")
a=say()
print(a)
```

```python
#매개변수를 지정하여 호출

def sub(a,b):
	result=a-b
	return result

c=sub(a=7,b=3)
print(c)
```
