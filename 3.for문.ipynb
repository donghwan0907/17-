```python
testlist=["one","two","three"]

for tl in testlist:
	print(tl)
	print(len(tl))
```

```python
a=[(1,2),(3,4),(5,6)]
for (t) in a:
	print(t)
print("-----------")
a=[(1,2),(3,4),(5,6)]
for (k,j) in a:
	print(j)
print("-----------")
for (k,j) in a:
	print(k)
```

```python
marks=[90,25,67,45,80]

num=0
for m in marks:
	num =num+1
	if m >=60:
		print(num,"학생은 합격입니다.")
	else:
		print(num,"학생은 불합격입니다.")
```

```python
#for문과 continue
marks=[90,25,67,45,80]
num=0
for m in marks:
	num+=1
	if m <60:
		continue
	print(num,"학생은 합격입니다.")
```

```python
#for와 range를 이용한 구구단
for i in range(2,4):
	for j in range(1,10):
		print(i,"X",j,"=",i*j)
```
