```python
#딕셔너리는 어떻게 만들까?
#{key1: value1, key2: value2, key3: value3}
dic={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(dic)
print(type(dic))
```

```python
#딕셔너리 쌍 추가하기
a={1: 'a'}
a[2]='b' #[]안은 index가 아니라 key값을 넣어야 한다.
print(a)
print(type(a))
a[3]=[1,2,3]
print(a)
a["이름"]='pey'
print(a)
#딕셔너리 쌍 삭제하기
del a["이름"]
print(a)
```

```python
#딕셔너리 요소 개수
print(len(a))
```

```python
#딕셔너리에 key를 사용해서 value 얻기
grade={
	'pey': 10,
	'julliet': 99
}
print(grade['pey'])
print(a[3])
```

```python
#딕셔너리 관련 함수들

##1. key, value 리스트 만들기 (keys(), values())
dic={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(list(dic.keys()))
print(list(dic.values()))
```

```python
##2. key, value 쌍 얻기 (items())
dic={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(list(dic.items())) #하나의 요소는 튜플로 구성되어 있다.
```

```python
##3. key와 value 쌍 모두 지우기 (clear())
자기소개={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(자기소개)
자기소개.clear()
print(자기소개)
```

```python
##4. key로 value 얻기 (get())
dic={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print(dic.get('name'))
print(dic['name']) #get()과 []의 차이점은 없는 key를 호출했을 때 []는 오류가 발생하지만 get()은 None을 반환한다.
print(dic.get('hobby')) #없는 key를 호출했을 때 None을 반환한다.
```

```python
##5. 해당 key가 딕셔너리 안에 있는지 조사하기 (in)
dic={
	'name' : 'pey',
	'phone' : '010-1234-1234',
	'birth' : '1118'
}
print('name' in dic) # True
print('hobby' in dic) # False
```
